```python
# libraries to import
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# data source: https://www.kaggle.com/datasets/souvikahmed071/social-media-and-mental-health
smmh = pd.read_csv('smmh.csv')
smmh
```

```python
smmh.describe().transpose()
```

```python
# check for missing values
smmh.isnull().sum()
# there are no missing values
```

```python
# rename columns
print(smmh.columns)
```

```python
# copy of the dataset
smmh_copy = smmh.copy()
smmh_copy.set_axis(['timestamp',
                    'age', 
                    'gender',
                    'rel_stat', 
                   'occ_stat',
                    'org_affil',
                    'social_media',
                   'social_media_platforms',
                    'avg_time', 
                   'social_media_spec_purpose',
                    'social_media_dist', 
                   'social_media_restless',
                    'easily_dist',
                    'worries',
                   'conc_diff',
                    'social_media_comparison',
                    'comparison_feeling',
                   'social_media_validation',
                    'depressed',
                    'interes_fluc',
                   'sleep_issues'], axis='columns', inplace=True)
```

```python
# after renaming columns
    # column breakdown:
    # 1. timestamp: entry of the respondent
    # 2. age: age of the respondent
    # 3. gender: respondent's gender
    # 4. rel_stat: respondent's relationship status
    # 5. occ_stat: resondent's occupation status
    # 6. org_affil: type of organizations respondent is affiliated with
    # 7. social_media: whether they respondent uses social media 
    # 8. social_media_platforms: social media platforms they commonly use
    # 9. avg_time: average time spent on social media daily
    # 10. social_media_spec_purpose: how often respondent finds themself using social media w/o a specific purpose
    # 11. social_media_dist: how often respondent gets distracted by social media while they are busy doing something
    # 12. social_media_restless: whether or not respondent feels restless if they haven't used social media in a while
    # 13. easily_dist: how easily respondent is distracted (scale 1-5)
    # 14. worries: how bother respondent is by worries (scale 1-5)
    # 15. conc_diff: whether respondent finds it difficult to concentrate on things
    # 16. social_media_comparison: how often does respondent compare themself to other successful people via social media (scale 1-5)
    # 17. comparison_feeling: how do you feel about these comparisons
    # 18.  social_media_validation: how often do you look to seek validation from features of social media
    # 19. depressed: how often do you feel depressed or down
    # 20. interes_fluc: how frequently does your interest in daily activities fluctuate (scale 1-5)
    # 21. sleep_issues: how often do you face issues regarding sleep (scale 1-5)
smmh_copy
```

```python
# how many people have social media?
smmh_copy['social_media'].value_counts()
```

```python
# social media is a categorical value. we can change this
social_media_dict = {
    'No': 0,
    'Yes': 1
}
smmh_copy['social_media'] = smmh_copy['social_media'].map(social_media_dict)
smmh_copy.head()

# now values are changed from categorical to quantitative
```
